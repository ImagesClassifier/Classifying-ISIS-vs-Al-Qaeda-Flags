# Data Augmentation

Data augmentation occurs when you create new data based on modifications of existing data. In our case, the data we'll work with will be images. Data augmentation on images would include transformations like:
   * Flipping the image either horizontally or vertically,
   * Rotating the image
   * Zooming in or out the image
   * Cropping the image
   * Varying the color on the image

```python
import matplotlib.pyplot as plt
from scipy import ndimage
import numpy as np
import pandas as pd
import os
import shutil
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import itertools
import shutil
from random import randint
import random
import glob
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
%matplotlib inline
```

```python
gen = ImageDataGenerator(rotation_range=90,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.15,
    zoom_range=0.1, 
    horizontal_flip=True)
```

```python
image_path = 'C:\\Users\\רועי\\images\\isis.png'
```

```python
image = np.expand_dims(plt.imread(image_path), 0)
```

```python
plt.imshow(image[0])
```

```python
save_here = 'C:\\Users\\רועי\\images\\isis'
```

```python
gen.fit(image)
```

```python
for x, val in zip(gen.flow(image,\
                           save_to_dir=save_here,\
                           save_prefix='isis',\
                           save_format='png'),range(700)) :
 pass
```
